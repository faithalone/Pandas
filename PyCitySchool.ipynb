```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_file_path ="C:/Users/remna/Desktop/Pandas_Assignment/04-Pandas/Homework/PyCitySchools/Resources/schools_complete.csv"
student_file_path ="C:/Users/remna/Desktop/Pandas_Assignment/04-Pandas/Homework/PyCitySchools/Resources/students_complete.csv"

# Read the csv files
school_data = pd.read_csv(school_file_path)

student_data = pd.read_csv(student_file_path)
student_data = pd.DataFrame(student_data)
```

```python
# Total Number of Schools
Total_schools = school_data["School ID"].count()
Total_schools
```

```python
# Total number of students
Total_students = school_data["size"].sum()
Total_students
```

```python
# Calculated the total budget
Total_budget = school_data["budget"].sum()
Total_budget
```

```python
#Calculate the average math score
Average_math_score = student_data["math_score"].mean()
Average_math_score
```

```python
Average_reading_score = student_data["reading_score"].mean()
Average_reading_score
```

```python
# Percentage of students with a passing math score (70 or greater)
Passing_math_perc = (student_data.loc[student_data["math_score"]>=70]).count()["student_name"]
Passing_math_perc= Passing_math_perc / Total_students * 100
Passing_math_perc
```

```python
# Percentage of students with a passing reading score (70 or greater)
Passing_reading_perc = (student_data.loc[student_data["reading_score"]>=70]).count()["student_name"]
Passing_reading_perc= Passing_reading_perc / Total_students * 100
Passing_reading_perc
```

```python
# Overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
Overall_Ave_score = (Passing_reading_perc + Passing_math_perc) / 2
Overall_Ave_score
```

```python
# Create a dataframe to hold the above results
District_Summary = pd.DataFrame({"Total Schools": [Total_schools],
                                 "Total Students": [Total_students],
                                 "Total Budgets": [Total_budget],
                                 "Average Math Score": [Average_math_score],
                                 "Average Reading Score": [Average_reading_score],
                                 "% Passing Math": [Passing_math_perc],
                                 "% Passing Reading": [Passing_reading_perc],
                                 "Overall Passing Rate": [Overall_Ave_score]})
# Use Map to format all the columns
District_Summary["Average Math Score"] = District_Summary["Average Math Score"].map("{:.2f}".format)
District_Summary["Average Reading Score"] = District_Summary["Average Reading Score"].map("{:.2f}".format)
District_Summary["% Passing Math"] = District_Summary["% Passing Math"].map("{:.2f}".format)
District_Summary["% Passing Reading"] = District_Summary["% Passing Reading"].map("{:.2f}".format)
District_Summary["Overall Passing Rate"] = District_Summary["Overall Passing Rate"].map("{:.2f}".format)



District_Summary
```
